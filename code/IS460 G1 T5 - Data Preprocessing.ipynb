# 2. Feature Selection

<hr>
<i>Back to <a href='#Table-of-Contents'>Table of Contents</a></i>

```python
# dropping Unnamed column as it merely describes row index
df = df.drop('Unnamed: 0', axis=1)
df.head()
```

<br>
As the dataset contains two date attributes, <strong>Review Date</strong> and <strong>Date Flown</strong>, only the latter will be considered as it provides a more precise indication of the date of the flight.
<br>

```python
# dropping review date column due to feature redundancy
df.drop(columns=['Review Date'], inplace=True)
df.head()
```

```python
# dropping attributes due to feature irrelevance
df.drop(columns=['Airline Name', 'Verified', 'Aircraft', 'Type Of Traveller', 'Seat Type', 'Route'], inplace=True)
df.head()
        
## note: review title and review are left for sentiment analysis / llm application later, but will not be addressed in this section
```

##

# 3. Segmentation by Time Period
<strong>Pre-COVID period:</strong> Before 1 March 2019 (exclusive)<br>
<strong>Post-COVID period:</strong> After 1 March 2019 (inclusive)
<hr>
<i>Back to <a href='#Table-of-Contents'>Table of Contents</a></i>

```python
# converting review date attribute into datetime object
df['Date Flown'] = pd.to_datetime(df['Date Flown'], format='%B %Y')
df.head()
```

```python
print(df['Date Flown'].dtype)
```

```python
# cutoff date for pre- and post-COVID periods
covid_cutoff = pd.Timestamp('2020-03-01')

# creating pre- and post-COVID DataFrames
pre_covid = df[df['Date Flown'] < covid_cutoff]
post_covid = df[df['Date Flown'] >= covid_cutoff]
```

```python
# checking most recent date of pre-covid data
pre_covid['Date Flown'].head()
```

```python
# checking start of post-covid data
post_covid['Date Flown'].tail()
```

# 4. Addressing Missing Values
We address missing values after segmenting the data to <strong>avoid information leakage</strong> between the time periods. Furthermore, as GAN algorithms require <strong>complete data</strong>, imputation must prelude applying the GAN algorithm.
<hr>
<i>Back to <a href='#Table-of-Contents'>Table of Contents</a></i>

## Imputing Pre-Covid DataFrame

```python
# descriptive statistics BEFORE imputation
pre_covid.describe()
```

```python
# retrieving the count of null values for each column
null_counts = pre_covid.isnull().sum()
print(null_counts)
```

```python
from sklearn.impute import SimpleImputer

imputer = SimpleImputer(strategy='median')
pre_covid['Seat Comfort'] = imputer.fit_transform(pre_covid[['Seat Comfort']])
pre_covid['Cabin Staff Service'] = imputer.fit_transform(pre_covid[['Cabin Staff Service']])
pre_covid['Food & Beverages'] = imputer.fit_transform(pre_covid[['Food & Beverages']])
pre_covid['Ground Service'] = imputer.fit_transform(pre_covid[['Ground Service']])
pre_covid['Inflight Entertainment'] = imputer.fit_transform(pre_covid[['Inflight Entertainment']])
pre_covid['Wifi & Connectivity'] = imputer.fit_transform(pre_covid[['Wifi & Connectivity']])

null_counts = pre_covid.isnull().sum()
print(null_counts)
```

```python
# descriptive statistics AFTER imputation
pre_covid.describe()
```

<hr>

## Imputing Post-Covid DataFrame

```python
# descriptive statistics BEFORE imputation
post_covid.describe()
```

```python
# retrieving the count of null values for each column
null_counts = post_covid.isnull().sum()
print(null_counts)
```

```python
post_covid['Seat Comfort'] = imputer.fit_transform(post_covid[['Seat Comfort']])
post_covid['Cabin Staff Service'] = imputer.fit_transform(post_covid[['Cabin Staff Service']])
post_covid['Food & Beverages'] = imputer.fit_transform(post_covid[['Food & Beverages']])
post_covid['Ground Service'] = imputer.fit_transform(post_covid[['Ground Service']])
post_covid['Inflight Entertainment'] = imputer.fit_transform(post_covid[['Inflight Entertainment']])
post_covid['Wifi & Connectivity'] = imputer.fit_transform(post_covid[['Wifi & Connectivity']])
post_covid['Value For Money'] = imputer.fit_transform(post_covid[['Value For Money']])

null_counts = post_covid.isnull().sum()
print(null_counts)
```

```python
# descriptive statistics AFTER imputation
post_covid.describe()
```

# 5. Generative Adversarial Network (GAN)
In our initial analysis of the <a href='https://www.kaggle.com/code/gvyshnya/airline-reviews-sentiments-ts-and-eda/input'>data source</a>, we discovered that there is a massive data imbalance with respect to the pre- and post-COVID classes. This means that, if we use the data as-is, our models' <strong>learning capacity for the pre-COVID period is lesser</strong> compared to that of the post-COVID period. 

Thus, we intend to utilise the GAN algorithm to:
<li>Generate more airline review data for the post-COVID period by training it on the limited amount data pertaining to this time period</li>
<li>Fine-tuning GAN to simulate the distribution of post-COVID reviews.</li>
<hr>
<i>Back to <a href='#Table-of-Contents'>Table of Contents</a></i>

###

```python
# balancing the data by generating more samples for the post-COVID period using GAN
print(f"Pre-COVID data: {len(pre_covid)}\nPost-COVID data: {len(post_covid)}")
```

<br>In view of the amount of data in the <strong>pre-COVID</strong> period being less than that of the post-covid period, we will be using GAN on the pre-covid data segment only.

```python
pre_covid.head()
```

```python
# encoded recommended as it is a binary categorical attribute
from sklearn.preprocessing import MinMaxScaler, LabelEncoder

label_encoder = LabelEncoder()
pre_covid['Recommended'] = label_encoder.fit_transform(pre_covid['Recommended'])  # 'yes' -> 1, 'no' -> 0
pre_covid.head()
```

```python
# exporting data
pre_covid.to_csv('pre_covid.csv', index=False)
post_covid.to_csv('post_covid.csv', index=False)
```
