```python
# Import pandas to handle and manipulate data as DataFrames
import pandas as pd

# Import numpy for numerical operations and array manipulation
import numpy as np

# Import matplotlib's pyplot for data visualization (e.g., plotting graphs)
import matplotlib.pyplot as plt

# Import TensorFlow, a deep learning framework for building neural networks
import tensorflow as tf

# Import Keras, a high-level API that runs on top of TensorFlow for easier neural network creation
import keras

# Import train_test_split from scikit-learn to split the dataset into training and testing sets
from sklearn.model_selection import train_test_split

# Import LabelEncoder from scikit-learn to convert categorical labels into numeric values
from sklearn.preprocessing import LabelEncoder, StandardScaler

# Import accuracy_score, confusion_matrix, and classification_report from sklearn
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

# Import the Sequential model from Keras, which allows us to build a feed-forward neural network layer by layer
from keras.models import Sequential

# Import Dense from Keras, which adds fully connected layers to the neural network
from keras.layers import Dense
```

```python
# Load the pre-COVID dataset into a DataFrame
pre_covid_df = pd.read_csv('/content/drive/MyDrive/smu projects/is460 machine learning & applications/pre_covid_new.csv')

# Load the post-COVID dataset into a separate DataFrame
post_covid_df = pd.read_csv('/content/drive/MyDrive/smu projects/is460 machine learning & applications/post_covid.csv')
```

## 2.1. Pre-Covid Analysis

<hr>
<i>Back to <a href='#Table-of-Contents'>Table of Contents</a></i>

```python
# Step 1: Select the features (X) from the pre-COVID dataset
X = pre_covid_df[['Overall_Rating', 'Seat Comfort', 'Cabin Staff Service', 'Food & Beverages',
                  'Ground Service', 'Inflight Entertainment', 'Wifi & Connectivity', 'Value For Money']]

# Step 2: Select the target variable (y) from the pre-COVID dataset (Recommended column)
# Convert 'yes' to 1 and 'no' to 0 for binary classification in the target variable
y = pre_covid_df['Recommended'].map({'yes': 1, 'no': 0})

# Step 3: Initialize the StandardScaler to standardize the features in X (to have mean=0 and std=1)
scaler = StandardScaler()

# Fit the scaler to the data (calculate mean and std), then transform X to the scaled values
X_scaled = scaler.fit_transform(X)

# Step 4: Split the data into training and testing sets (80% for training, 20% for testing)
# This helps evaluate the model on unseen data
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=20)
```

```python
# Step 1: Initialize a Sequential model
# The Sequential model allows you to stack layers one after another
model1 = Sequential()

# Step 2: Add the first input layer with 64 neurons
# 'input_dim' specifies the number of features in X_train, and 'relu' is the activation function
model1.add(Dense(64, input_dim=X_train.shape[1], activation='relu'))

# Step 3: Add a second hidden layer with 32 neurons
# 'relu' (Rectified Linear Unit) activation is used for non-linearity
model1.add(Dense(32, activation='relu'))

# Step 4: Add the output layer with 1 neuron
# 'sigmoid' activation is used for binary classification (output values between 0 and 1)
model1.add(Dense(1, activation='sigmoid'))

# Step 5: Compile the model
# 'binary_crossentropy' is used as the loss function for binary classification
# 'adam' optimizer is used for efficient gradient descent
# 'accuracy' is the evaluation metric during training
model1.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

# Step 6: Train the model on the training data (X_train, y_train)
# Run for 100 epochs, with a batch size of 32, and validate the model on test data (X_test, y_test)
model1.fit(X_train, y_train, epochs=100, batch_size=32, validation_data=(X_test, y_test))
```

```python
# Step 1: Make predictions on the test data using the trained model
# The predictions are probabilities (continuous values between 0 and 1)
y1_pred = model1.predict(X_test)

# Step 2: Convert the predicted probabilities into binary values (1 or 0)
# If the probability is greater than 0.5, classify as 1; otherwise, classify as 0
y1_pred_binary = (y1_pred > 0.5).astype(int)

# Step 3: Calculate the accuracy of the model (proportion of correct predictions)
accuracy = accuracy_score(y_test, y1_pred_binary)

# Step 4: Calculate the classification error (1 - accuracy), representing the misclassification rate
classification_error = 1 - accuracy

# Step 5: Generate a confusion matrix to visualize true positives, true negatives, false positives, and false negatives
conf_matrix = confusion_matrix(y_test, y1_pred_binary)

# Step 6: Extract the True Negatives (tn), False Positives (fp), False Negatives (fn), and True Positives (tp) from the confusion matrix
tn, fp, fn, tp = conf_matrix.ravel()

# Step 7: Calculate precision: the proportion of positive predictions that were correct
precision = tp / (fp + tp)

# Step 8: Calculate recall (True Positive Rate): the proportion of actual positives that were correctly identified
recall = tp / (fn + tp)

# Step 9: Calculate the F1-Score: the harmonic mean of precision and recall
f1_score = (recall * precision) * 2 / (precision + recall)

# Step 10: Calculate specificity (True Negative Rate): the proportion of actual negatives that were correctly identified
specificity = tn / (fp + tn)

# Step 11: Print the model's accuracy
print("Accuracy:", accuracy)

# Step 12: Print the classification error (misclassification rate)
print("Classification Error:", classification_error)

# Step 13: Print the precision score
print("Precision:", precision)

# Step 14: Print the recall (sensitivity) score
print("Recall:", recall)

# Step 15: Print the F1-Score
print("F1 Score:", f1_score)

# Step 16: Print the specificity (true negative rate)
print("Specificity:", specificity)

# Step 17: Generate and print a detailed classification report (including precision, recall, F1-score, and support for each class)
report = classification_report(y_test, y1_pred_binary)
print("\nClassification Report Post-COVID dataset:\n", report)
```

```python
from tensorflow.keras.utils import plot_model

plot_model(model1, show_shapes=True)
```
