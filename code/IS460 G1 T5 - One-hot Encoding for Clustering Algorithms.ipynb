```python
import pandas as pd

# Load the dataset
data = pd.read_csv("Airline_review.csv")

# Display the first few rows of the dataset
print("Original Data:")
print(data.head())

# One-hot encoding for the specified categorical variables
data_encoded = pd.get_dummies(data, columns=["Type Of Traveller", "Seat Type", "Recommended"], drop_first=True)

# Convert boolean columns (True/False) to integers (1/0)
data_encoded = data_encoded.applymap(lambda x: 1 if x is True else (0 if x is False else x))

# Display the first few rows of the encoded dataset
print("\nEncoded Data:")
print(data_encoded.head())

# Optionally, save the encoded dataset to a new CSV file
data_encoded.to_csv("Airline_review_encoded.csv", index=False)

```
